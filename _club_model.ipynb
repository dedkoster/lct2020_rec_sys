```python
%%time
df = pd.read_csv('data_hobbies_common.csv', sep=';')

cols_items = [c for c in df.columns if c[:3] in ['m__', 'sch', 'gr_']]
cols_users = ['возраст', 'sex__Женский', 'sex__Мужской']
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

```python
df.loc[df.shape[0], :] = [1,-1] + [0]*(len(df.columns) - 2)
# df.tail()
```

```python
# default number of recommendations
K = 20
# model learning rate
LEARNING_RATE = 0.25
# no of latent factors
NO_COMPONENTS = 20
# no of epochs to fit model
NO_EPOCHS = 20

# regularisation for both user and item features
ITEM_ALPHA=1e-5
USER_ALPHA=1e-5

# seed for pseudonumber generations
SEEDNO = 321
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import lightfm
from lightfm import LightFM
from lightfm.data import Dataset
from lightfm import cross_validation

from lightfm.evaluation import precision_at_k as lightfm_prec_at_k
from lightfm.evaluation import recall_at_k as lightfm_recall_at_k
```

```python
import gc
gc.collect()
```

```python
from scipy.sparse import hstack, csr_matrix
```

```python
%%time
ii = csr_matrix(df[['id_услуги']+cols_items].values)
ii
```

```python
user_feat = csr_matrix(df[['id_ученика']+cols_users].values)
user_feat
```

```python
%%time
dataset = Dataset()
dataset.fit(df['id_ученика'].unique(), 
            df['id_услуги'].unique())#, 
            #item_features=ii,#item_feat,
            #user_features=user_feat)
```

```python
%%time
(interactions, weights) = dataset.build_interactions(df[['id_ученика','id_услуги']].drop_duplicates().values)
```

```python
interactions
```

```python
uid_map, _, iid_map, _ = dataset.mapping()
```

```python
%%time
joblib.dump(interactions, open('club_interactions.pkl','wb'))
joblib.dump(weights, open('club_weights.pkl','wb'))
joblib.dump(uid_map, open('club_uid_map.pkl','wb'))
joblib.dump(iid_map, open('club_iid_map.pkl','wb'))
```

```python
%%time
joblib.dump(ii, open('hobbie_item_feat.pkl','wb'))
joblib.dump(user_feat, open('hobbie_user_feat.pkl','wb'))
```

```python
model = LightFM(loss='warp', no_components=NO_COMPONENTS, 
                 learning_rate=LEARNING_RATE, 
                 item_alpha=ITEM_ALPHA,
                 user_alpha=USER_ALPHA,
                 random_state=np.random.RandomState(SEEDNO))
```

```python
# model = joblib.load(open('model_book01.pkl','rb'))
```

```python
%%time
model.fit(interactions=interactions, verbose=True,
           user_features=user_feat,
           item_features=ii,
           epochs=NO_EPOCHS)
```
